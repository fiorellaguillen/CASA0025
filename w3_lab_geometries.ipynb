```python
%%sql

CREATE TABLE IF NOT EXISTS neighborhoods AS
SELECT *
FROM nyc_neighborhoods;

FROM neighborhoods;
```

**EXERCISE 1**

```python
%%sql

SELECT ST_AREA(geom)/10000 AS 'area in ha'
FROM neighborhoods
WHERE NAME= 'West Village'
```

**EXERCISE 2**

```python
%%sql

CREATE TABLE IF NOT EXISTS streets AS
SELECT *
FROM nyc_streets;

FROM streets LIMIT 5
```

```python
%%sql

SELECT ST_GEOMETRYTYPE(geom) AS 'geom type', ST_LENGTH(geom) AS 'geom length'
FROM streets
WHERE NAME= 'Pelham St'
```
